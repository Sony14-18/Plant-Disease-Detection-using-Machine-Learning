```python
import tensorflow as tf

keras_model = tf.keras.models.load_model('/content/drive/MyDrive/miniproject/plant1.h5')
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(keras_model)
tflite_model = converter.convert()
```

```python
with tf.io.gfile.GFile('/content/drive/MyDrive/miniproject/plant1.tflite', 'wb') as f:
    f.write(tflite_model)
```
